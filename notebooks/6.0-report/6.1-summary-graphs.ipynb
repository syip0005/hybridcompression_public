```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
summary_folder = "../../summary_results/"
```

```python
cer_12days = pd.read_csv(summary_folder + 'cer_12days_model_comparison.csv')
umass_12days = pd.read_csv(summary_folder + 'umass_12days_model_comparison.csv')
dred_60rows = pd.read_csv(summary_folder + 'dred_60days_model_comparison.csv')
alternative_12days = pd.read_csv(summary_folder + 'alternative_12days.csv')
inputdays = pd.read_csv(summary_folder + 'input_days_comparison.csv')
```

### CER (12 days)

```python
cer_12days[cer_12days.latent_dimension == 48][['model', 'best_val_loss']]
```

```python
fig, ax = plt.subplots(figsize=(10,6))

alt_ = alternative_12days[alternative_12days.dataset == 'cer']
dl_ = cer_12days[['model', 'cr', 'best_val_loss']].rename({'best_val_loss':'mse'}, axis=1)
dl_.loc[dl_.model == 'HAE', 'model'] = 'HAE (Proposed)'
dl_.loc[dl_.model == 'SCSAE', 'model'] = 'SCSAE (Previous SoA)'
all_ = pd.concat([alt_, dl_]).reset_index()
all_ = all_[(all_.cr <= 20) & (all_.cr >= 2)]

sns.lineplot(data = all_, x = 'cr',  y = 'mse', hue = 'model', ax=ax, marker = 'o')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.invert_xaxis()

ax.set_title('cern val, num_days = 12')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

alt_ = alternative_12days[alternative_12days.dataset == 'cer']
dl_ = cer_12days[['model', 'cr', 'train_loss']].rename({'train_loss':'mse'}, axis=1)
dl_.loc[dl_.model == 'HAE', 'model'] = 'HAE (Proposed)'
dl_.loc[dl_.model == 'SCSAE', 'model'] = 'SCSAE (Previous SoA)'
all_ = pd.concat([alt_, dl_]).reset_index()
all_ = all_[(all_.cr <= 20) & (all_.cr >= 2)]

sns.lineplot(data = all_, x = 'cr',  y = 'mse', hue = 'model', ax=ax, marker = 'o')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

ax.invert_xaxis()

ax.set_title('cern train, num_days = 12')
plt.show()
```

### Input Days Comparison (Maintain CR = 12 or 18)

#### CER

```python
fig, ax = plt.subplots()

cer_input_days_ = inputdays[(inputdays.cr == 12) & (inputdays.dataset == "CERN3")]
cer_input_days_ = pd.melt(cer_input_days_, 
                          id_vars = ['ae_input_days', 'latent_dimension'], 
                          value_vars = ['best_val_loss', 'train_loss'])

sns.scatterplot(data = cer_input_days_, x = 'ae_input_days', y='value', 
             hue='variable', marker='o', ax=ax, size='latent_dimension')

sns.lineplot(data = cer_input_days_, x = 'ae_input_days', y='value', 
             hue='variable', ax=ax, legend=False)


box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```
